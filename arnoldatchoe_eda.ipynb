## STEP 6: INITIAL DISTRIBUTION OF CATEGORICAL COLUMNS
We check the distribution of our categorical features or columns among unique categories. We can use the first line of the code box below to check distribution of specific columns or check distribution for all columns simultaneously using the second line of the code box below.

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

## OBSERVATION 2:CATEGORICAL COLUMN DISTRIBUTION
From the histogram above, we can observe that the number of iris flowers is equal across the different species.

## STEP 7: DATA TRANSFORMATION AND ENGINEERING
In this sept, we rename the sepal_length column and also created a new column called Sepal_Area, a result of the product of Sepal_Length and sepal_width.


```python
# Renaming columns to include unit of measurement for width and length
df.rename(columns={'sepal_length': 'sepal_length_cm'}, inplace=True)
df.rename(columns={'sepal_width': 'sepal_width_cm'}, inplace=True)
df.rename(columns={'petal_length': 'petal_length_cm'}, inplace=True)
df.rename(columns={'petal_width': 'petal_width_cm'}, inplace=True)

# Adding a new column
df['sepal_Area'] = df['sepal_length_cm'] * df['sepal_width_cm']
df['petal_Area'] = df['petal_length_cm'] * df['petal_width_cm']
```

```python
# Checking the relationship between various numerical variables.

# Create scatter plots to examine the relationship between numerical columns or features of your DataFrame
plt = sns.pairplot(data=df, hue="species")

# Run this cell by clicking the play button (Execute cell and below) in the cell toolbar
# Or by pressing Shift+Enter 
# or by clicking Run All in the Menu above
```

3. Setosa has low petal width and lenght.
4. Versicolor has moderate or average width and length.
